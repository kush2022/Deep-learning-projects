# Data augmentation 
- used to address overfitting and improve the generalization of models. 
- It involves creating new training examples by applying various transformations and perturbations to the original training data. 
- These transformaations introduce diversity into the dataset, making the model more robust and reducing its tendency to overfit. 


**Here are some common data augmentation techniques and their applications:**

## Image Data Augmentation:
- **Rotation:** Rotate the image by a random angle to account for variations in object orientations.
- **Translation:** Shift the image horizontally and vertically to simulate different object positions within the frame.
- **Scaling:** Resize the image to different scales to account for variations in object size.
- **Scaling:** Resize the image to different scales to account for variations in object size.**
- **Shearing:** Apply a shearing transformation to simulate affine distortions.
- **Zooming:** Zoom in or out on the image to capture different levels of detail.
- **Color Jittering:** Modify brightness, contrast, saturation, and hue to account for changes in lighting conditions.


```python
import tensorflow as tf 
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt 
```

```python
# load the dataset 
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# create an imagedatagenerator for data augmentation 
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)


# choose an image from the dataset 
img = X_train[0]
img
```

```python
# reshape 
img = img.reshape((1,)+ img.shape)
img
```

```python
# Generate augmented images and display them
i = 0
for batch in datagen.flow(img, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(batch[0])
    i += 1
    if i % 4 == 0:
        break

plt.show()
```
